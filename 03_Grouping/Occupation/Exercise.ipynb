### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv(
    "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",
    sep="|"
)

users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation").age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def count_people(gender: str):
    if gender == "M":
        return 1
    else:
        return 0


users["gender_count"] = users.gender.apply(count_people)
(users.groupby("occupation")["gender_count"].sum() / users.groupby("occupation")[
    "gender_count"
].count()).sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation").age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"]).age.mean().reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
def pr(x: pd.Series):
    print(x.loc[:,(["gender_count"], ["count", "sum"])])


counts = users.groupby(["occupation"])[["gender_count"]].agg(
    {"gender_count": ["sum", "count"]}
)

counts.apply(pr)
# counts.loc[:, (["gender_count"], ["count", "sum"])]
```
